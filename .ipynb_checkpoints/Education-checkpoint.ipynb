## Abstract
The outbreak of the COVID-19 pandemic forced an unprecedented shift in education systems worldwide, leading to the widespread adoption of remote learning. This study aims to comprehensively investigate the impact of remote learning and the COVID-19 pandemic on student education outcomes. We are using a dataset comprising academic performance metrics before and after the pandemic pushed remote learning into action. By comparing pre-pandemic and pandemic-era data, we aim to analyze the specific effects of the transition to remote learning on academic achievement and overall educational attainment. Ultimately, this research endeavors to inform policymakers, educators, and stakeholders about the challenges and opportunities presented by the transition to remote learning in the context of the COVID-19 pandemic.

## About 
We utilized three datasets for our analysis. Two of these datasets provide insights into the distribution of remote, in-person, and hybrid schooling models within each state, with one dataset capturing data from 2020 and the other from 2021. The third dataset focuses on quantitative SAT statistics across states, encompassing data from both 2020 and 2021. An aspect we need to consider is that certain states, such as North Dakota, exhibit a preference for the ACT over the SAT. However, this discrepancy does not significantly affect our analysis, as our focus lies on assessing changes in state SAT performance rather than solely on academic achievement.

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings  # Import warnings module to handle warnings
from sklearn.cluster import KMeans

# Load the data
sat20 = pd.read_csv('sat20.csv')
sat21 = pd.read_csv('sat21.csv')

# Drop the first row (US Average) and unnecessary columns
sat20 = sat20.iloc[1:, [0, 3, 5]]
sat21 = sat21.iloc[1:, [0, 3, 5]]

# Rename the columns
sat20.columns = ['State', 'Mean of ERW', 'Mean of Math']
sat21.columns = ['State', 'Mean of ERW', 'Mean of Math']

# Function to calculate WCSS for a range of cluster numbers
def calculate_wcss(data, max_clusters=10):
    wcss = []
    for i in range(1, max_clusters+1):
        kmeans = KMeans(n_clusters=i, random_state=42)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)
    return wcss

# Calculate WCSS for 2020 data
X_2020 = sat20[['Mean of ERW', 'Mean of Math']].values

# Suppress FutureWarning for KMeans default value change
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    wcss_2020 = calculate_wcss(X_2020)

# Calculate WCSS for 2021 data
X_2021 = sat21[['Mean of ERW', 'Mean of Math']].values

# Suppress FutureWarning for KMeans default value change
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    wcss_2021 = calculate_wcss(X_2021)

# Plot the elbow method results
plt.figure(figsize=(10, 5))

# 2020 elbow plot
plt.subplot(1, 2, 1)
plt.plot(range(1, len(wcss_2020) + 1), wcss_2020, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters (2020)')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')

# 2021 elbow plot
plt.subplot(1, 2, 2)
plt.plot(range(1, len(wcss_2021) + 1), wcss_2021, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters (2021)')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import warnings

# Read the datasets
df1 = pd.read_csv('sat20.csv')
df2 = pd.read_csv('sat21.csv')

# Extract SAT scores for clustering
X_2020 = df1.iloc[1:, [1,2]].values  # Exclude 'States' column
X_2021 = df2.iloc[1:, [1,2]].values  # Exclude 'States' column

# Suppress FutureWarnings for KMeans default value changes
warnings.filterwarnings("ignore", category=FutureWarning)

# Perform k-means clustering
kmeans_2020 = KMeans(n_clusters=3, n_init=10, random_state=42)
kmeans_2021 = KMeans(n_clusters=3, n_init=10, random_state=42)

# Fit and predict clusters
clusters_2020 = kmeans_2020.fit_predict(X_2020)
clusters_2021 = kmeans_2021.fit_predict(X_2021)

# Plot clustering results for 2020 and 2021
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.scatter(X_2020[:, 0], X_2020[:, 1], c=clusters_2020, cmap='viridis', edgecolor='k', s=50)
plt.scatter(kmeans_2020.cluster_centers_[:, 0], kmeans_2020.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.title('K-Means Clustering of 2020 SAT Scores')
plt.xlabel('Mean of Total')
plt.ylabel('Mean of ERW')
plt.legend()

plt.subplot(2, 1, 2)
plt.scatter(X_2021[:, 0], X_2021[:, 1], c=clusters_2021, cmap='viridis', edgecolor='k', s=50)
plt.scatter(kmeans_2021.cluster_centers_[:, 0], kmeans_2021.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.title('K-Means Clustering of 2021 SAT Scores')
plt.xlabel('Mean of Total')
plt.ylabel('Mean of ERW')
plt.legend()

# Add lists of values in each cluster underneath the plots
plt.subplots_adjust(hspace=0.6)  # Increase the vertical spacing between subplots

# List of values in each cluster for 2020
print("\n2020 SAT Scores Clusters:")
for cluster_label in range(kmeans_2020.n_clusters):
    cluster_data = df1.iloc[1:][clusters_2020 == cluster_label]
    states_in_cluster = ", ".join(cluster_data['States'].values)
    print(f"Cluster {cluster_label}: {states_in_cluster}")

# List of values in each cluster for 2021
print("\n2021 SAT Scores Clusters:")
for cluster_label in range(kmeans_2021.n_clusters):
    cluster_data = df2.iloc[1:][clusters_2021 == cluster_label]
    states_in_cluster = ", ".join(cluster_data['States'].values)
    print(f"Cluster {cluster_label}: {states_in_cluster}")

plt.show()

```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import warnings

# Read the datasets
df1 = pd.read_csv('sat20.csv')
df2 = pd.read_csv('sat21.csv')


# Extract SAT scores for clustering
X_2020 = df1.iloc[1:, [3,5]].values  # Exclude 'States' column
X_2021 = df2.iloc[1:, [3,5]].values  # Exclude 'States' column

# Suppress FutureWarning for KMeans default value change
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    # Perform k-means clustering
    kmeans_2020 = KMeans(n_clusters=3, random_state=42)
    kmeans_2021 = KMeans(n_clusters=3, random_state=42)

    # Fit and predict clusters
    clusters_2020 = kmeans_2020.fit_predict(X_2020)
    clusters_2021 = kmeans_2021.fit_predict(X_2021)

# Plot clustering results for 2020
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_2020[:, 0], X_2020[:, 1], c=clusters_2020, cmap='viridis', edgecolor='k', s=50)
plt.scatter(kmeans_2020.cluster_centers_[:, 0], kmeans_2020.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.title('K-Means Clustering of 2020 SAT Scores')
plt.xlabel('Mean of Total')
plt.ylabel('Mean of ERW')
plt.legend()

# Plot clustering results for 2021
plt.subplot(1, 2, 2)
plt.scatter(X_2021[:, 0], X_2021[:, 1], c=clusters_2021, cmap='viridis', edgecolor='k', s=50)
plt.scatter(kmeans_2021.cluster_centers_[:, 0], kmeans_2021.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.title('K-Means Clustering of 2021 SAT Scores')
plt.xlabel('Mean of Total')
plt.ylabel('Mean of ERW')
plt.legend()

# List states in each cluster for 2020
plt.figure(figsize=(10, 5))
for i in range(3):
    cluster_i_states = df1['States'][1:][clusters_2020 == i].tolist()  # Start from index 1 to match X_2020 extraction
    print(f'Cluster {i+1} (2020): {", ".join(cluster_i_states)}')

# List states in each cluster for 2021
for i in range(3):
    cluster_i_states = df2['States'][1:][clusters_2021 == i].tolist()  # Start from index 1 to match X_2021 extraction
    print(f'Cluster {i+1} (2021): {", ".join(cluster_i_states)}')

plt.tight_layout()
plt.show()
 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load data from CSV files
sat_data = pd.read_csv('sat20.csv')
income_data = pd.read_csv('income.csv')

# Merge the datasets on the 'State' column
merged_data = pd.merge(sat_data, income_data, on='States')

# Select the relevant columns
X = merged_data[['Income', 'Mean of Total']]

# Standardize the data (optional but recommended)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform k-means clustering
k = 3  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)
labels = kmeans.labels_

# Visualize the clusters
plt.figure(figsize=(10, 6))

# Plot the data points
plt.scatter(X['Income'], X['Mean of Total'], c=labels, cmap='viridis', s=50, alpha=0.7)


# Add labels and title
plt.xlabel('Income')
plt.ylabel('Mean of Total SAT Scores')
plt.title('K-Means Clustering of SAT Scores and Income')
plt.legend()
plt.grid(True)
plt.show()

filtered_data['Cluster'] = labels

# Print the name of each state in each cluster
for cluster_num in range(k):
    states_in_cluster = filtered_data[filtered_data['Cluster'] == cluster_num]['States']
    print(f"Cluster {cluster_num + 1}:", ', '.join(states_in_cluster.tolist()))
```
